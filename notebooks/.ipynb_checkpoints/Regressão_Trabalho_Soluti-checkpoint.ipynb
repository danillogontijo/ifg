## Quantidade de dias desde a última data de suporte

```python
def diff_date(d1,d2):
    return d1-d2
```

```python
# dataset_merged = dataset_merged.sample(3000)
dataset_merged['data_suporte']= pd.to_datetime(dataset_merged['data_suporte'])
dataset_merged = dataset_merged.sort_values(by=['solicitacao_certificado', 'data_suporte'])

dataset_merged['diff_days_last_suport'] = dataset_merged.groupby(['solicitacao_certificado'])['data_suporte'].apply(lambda x: diff_date(x, x.shift().fillna(x)) )  #.apply(lambda x: x - x.iloc[0])
dataset_merged['diff_days_last_suport'] /= np.timedelta64(1,'D')
dataset_merged['diff_days_last_suport'] = dataset_merged['diff_days_last_suport'].astype(int)

# teste = teste.groupby(['solicitacao_certificado'])[['data_suporte','motivo_suporte']]
dataset_merged[dataset_merged['diff_days_last_suport'] > 0]
# teste.groupby(['solicitacao_certificado'])['data_suporte'].describe()

# dataset_merged
```

## Verifique a multicolinearidade usando VIF

- Ao contrário da matriz de correlação, VIF determina a força da correlação de uma variável com um grupo de outras variáveis independentes em um conjunto de dados. VIF começa geralmente em 1 e qualquer lugar que exceda 10 indica alta multicolinearidade entre as variáveis independentes.

```python
#Step 9.6: Check Multicolinearity using VIF----------------------------------------------

from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)

# dataset2 = dataset[['gender', 'SeniorCitizen', 'Partner', 'Dependents',
#        'tenure', 'PhoneService', 'PaperlessBilling','MonthlyCharges','TotalCharges']]

calc_vif(dataset2)
```
